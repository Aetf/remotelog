### Reason for latency variations

The variation comes from two factors.

- The execution time for `obj_track` and `streamer`
- The queuing time

And the root cause is that SlidingWindowBatcher can't handle frames with a small sequence number come in late after other frames with a near sequence number got processed. Thus the sliding window only moves when it reaches its max size. This can be seen in the latency for `obj_track`. The default max size is 24, which is why the period is around 24.

```python
p = otk_fps_rerun.exps[0].seq_latency('scale-obj_track')
p.set_xlim(240, 300)
p.set_ylim(0, 150)
```

Also, the latency reported for `obj_track` is the execution time for a whole batch, i.e. the latency for all frames inside a batch is the same. And for `obj_track` implementation, the batch execution time is linear to the batch size.

Due to the dynamic nature of the original SlidingWindow implementation, `obj_track` is getting different batch size, which depends on whether there is a missing frame. Thus the variation in execution time.

```python
p = otk_fps_rerun.exps[1].seq_latency('obj_track')
p.set_xlim(25, 91)
p.set_ylim(0, 70)
```
